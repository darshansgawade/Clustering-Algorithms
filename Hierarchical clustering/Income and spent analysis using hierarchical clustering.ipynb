# Income and Spent Analysis using Heirarchical Clustering

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
```

### Loading dataset

```python
data = pd.read_csv('dataset.csv')
```

```python
data
```

### Summarise the dataset

```python
data.shape
```

```python
data.columns
```

```python
data = data.drop('CustomerID',axis='columns')
```

```python
data
```

```python
data.describe()
```

```python
data.isna().sum()
```

### Label encoding

```python
data.replace(["Male","Female"],[0,1],inplace=True)
```

```python
data
```

### Building Model

```python
import scipy.cluster.hierarchy as clustering
```

```python
plt.figure(figsize=(18,6))
dendrogram= clustering.dendrogram(clustering.linkage(data,method='ward'))
plt.title("Dendrogram Tree Graph")
plt.xlabel("customers")
plt.ylabel("distance")
plt.show()
```

### Fitting Hierarchical clustering to dataset with n_clusters=5

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage="average")
y_means = model.fit_predict(data)
y_means
```

```python
x=data.iloc[:,[2,3]].values
```

```python
x
```

### Visualize the clusters

```python
plt.scatter(x[y_means==0,0],x[y_means==0,1],s=50,c="purple",label="Cluster 1")
plt.scatter(x[y_means==1,0],x[y_means==1,1],s=50,c="green",label="Cluster 2")
plt.scatter(x[y_means==2,0],x[y_means==2,1],s=50,c="yellow",label="Cluster 3")
plt.scatter(x[y_means==3,0],x[y_means==3,1],s=50,c="cyan",label="Cluster 4")
plt.scatter(x[y_means==4,0],x[y_means==4,1],s=50,c="red",label="Cluster 5")

plt.title("Income vs Annual Spending Graph")
plt.xlabel("Income")
plt.ylabel("Annual Spending")
plt.show()
```
